```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
netflix_py = pd.read_csv("data/netflix.csv")
```

```python
netflix_py
```

```python
netflix_py.groupby('MovieID')['Rating'].describe().T
```

```python
def describe(df):
    return pd.concat([df.groupby('MovieID')['Rating'].describe(),
                      df.groupby('MovieID')['Rating'].agg(pd.Series.mode).rename('mode'),
                      df.groupby('MovieID')['Rating'].agg(pd.Series.median).rename('median')
                     ], axis=1).T

describe(netflix_py)


```
